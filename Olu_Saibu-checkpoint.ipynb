```python
crimes_by_theft=crime_data_df.copy()
Pry_Types=['THEFT']
crimes_by_theft=crimes_by_theft[crimes_by_theft['Primary Type'].isin(Pry_Types)]
crimes_by_theft
```

```python
crime_by_theft_wordcloud = crimes_by_theft.copy()
crime_by_theft_wordcloud.dropna(axis=0, subset=["Block"], inplace=True)
# creating string of Block codes and USA flag image mask
crime_theft_locations =" ".join(crime for crime in crime_by_theft_wordcloud["Block"])
mask=np.array(Image.open(r"C:\Users\saibu\Documents\Bootcamp\Project\USA\United_States.jpg"))

# creating and generating a wordcloud image
crime_theft_location_wordcloud = WordCloud(background_color="white", mode="RGBA", max_words=5000, mask=mask).generate(crime_theft_locations)

# creating coloring from image and plot wordcloud
image_colors=ImageColorGenerator(mask)
plt.figure(figsize=[8,8])
plt.imshow(crime_theft_location_wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

plt.savefig(r'C:\Users\saibu\OneDrive\Desktop\project_image\fig2.png',transparent=True, bbox_inches='tight')
plt.show()
```

```python
crime_by_theft_wordcloud = crimes_by_theft.copy()
crime_by_theft_wordcloud.dropna(axis=0, subset=["Location Description"], inplace=True)
# creating string of Block codes and USA flag image mask
crime_theft_locations =" ".join(crime for crime in crime_by_theft_wordcloud["Location Description"])
# mask=np.array(Image.open("us.png"))
mask=np.array(Image.open(r"C:\Users\saibu\Documents\Bootcamp\Project\USA\United_States.jpg"))

# creating and generating a wordcloud image
crime_theft_location_wordcloud = WordCloud(background_color="white", mode="RGBA", max_words=800, mask=mask).generate(crime_theft_locations)

# creating coloring from image and plot wordcloud
image_colors=ImageColorGenerator(mask)
plt.figure(figsize=[8,8])
plt.imshow(crime_theft_location_wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

plt.savefig(r'C:\Users\saibu\OneDrive\Desktop\project_image\fig3.png',transparent=True, bbox_inches='tight')
plt.show()
```

```python
crimes_by_arrest= crimes_by_theft.copy()
crimes_by_arrest=crimes_by_arrest.groupby(['Primary Type','Arrest']).size().reset_index(name="Number of crimes")
crimes_by_arrest=crimes_by_arrest[crimes_by_arrest['Primary Type'].isin(Pry_Types)]
crimes_by_arrest_pivot=pd.pivot_table(crimes_by_arrest, index = ['Primary Type'],columns=['Arrest'], values='Number of crimes', aggfunc=np.sum)

crimes_by_arrest_pivot= crimes_by_arrest_pivot.reset_index()

crimes_by_arrest_pivot['Total Number']=crimes_by_arrest_pivot[True] + crimes_by_arrest_pivot[False]
crimes_by_arrest_pivot['True']=(crimes_by_arrest_pivot[True] / crimes_by_arrest_pivot['Total Number'])*100
crimes_by_arrest_pivot['False']=(crimes_by_arrest_pivot[False] / crimes_by_arrest_pivot['Total Number'])*100
crimes_by_arrest_pivot=crimes_by_arrest_pivot[['Primary Type','True','False']]
crimes_by_arrest_pivot=crimes_by_arrest_pivot.set_index('Primary Type')

plt.savefig(r'C:\Users\saibu\OneDrive\Desktop\project_image\figD.png',transparent=True, bbox_inches='tight')
crimes_by_arrest_pivot.plot(kind='barh', stacked=True, figsize=(8,3), title="Percentage of Theft Arrested/Not Arrested")
```

From the stacked column chart, abt 11.7% reportedlty arrested and about 88.3% not arrested(2001 to Sept 2019).
* CPD might want to look into better ways of tackling investigations.
* Possibility of false reports and allegations.
